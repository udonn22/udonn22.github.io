<h1>正規分布: 自然界と社会でのパターン</h1>
<h2>正規分布ってどんな分布?</h2>

<div class="speech-container">
<img class="child-pic" src="pictures\shibainu_2_1.png" alt="子供の顔画像">
<p class="child-txt">博士、正規分布ってよく聞くけど、それはどういう意味なの?</p>
</div>

<div class="speech-container">
<div class="hakase-txt">正規分布は、自然界や人間の行動においてよく見られるデータの分布のことだよ。例えば、人々の身長やテストの点数がこの分布に従っていることが多いんだ。</div>

<div class="speech-container">
<div class="hakase-txt">この分布は、下の図みたいにベルの形をしていて、データの多くが平均値の周りに集まり、極端に高い値や低い値は少なくなる特徴があるんだ。</div>
<img class="hakase-pic" src="pictures\shibainu_4_2.png">

```python
# !pip install japanize-matplotlib
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import japanize_matplotlib

# 正規分布のパラメータ
mu = 0  # 平均
sigma = 1  # 標準偏差

# 正規分布に従うデータの生成
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)
y = norm.pdf(x, mu, sigma)

# 正規分布のプロット
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='正規分布')
plt.title('正規分布 (μ=0, σ=1)')
plt.xlabel('X')
plt.ylabel('確率密度')
plt.legend()
plt.grid(True)
plt.show()
```

<link rel="stylesheet" type="text/css" href="style.css">
<div class="speech-container">
<div class="hakase-txt">ちなみに確率分布の式はこうだよ。</div>
<img class="hakase-pic" src="pictures\shibainu_4_2.png">
</div>

<div class="definition-container">
$$
f(x|\mu,\sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
$$
ただし、$x$は確率変数、$\mu$は平均（期待値）、$\sigma^2$は分散、$\sigma$は標準偏差です。
</div>


<div class="speech-container">
<div class="hakase-txt">確率変数$X$が正規分布に従うときは、以下のようにして書くことが多いから、覚えておくといいよ。</div>
<img class="hakase-pic" src="pictures\shibainu_4_3.png">
</div>

$$
X \sim N\left(\mu, \sigma^2\right)
$$

<h2>データのばらつきと平均をそろえる方法とは?</h2>
<div class="speech-container">
<img class="child-pic" src="pictures\shibainu_2_2.png" alt="子供の顔画像">
<p class="child-txt">どうやって標準正規分布に変換するの?</p>
</div>
<div class="speech-container">
<div class="hakase-txt">いい質問だね! データを標準正規分布に変換するには、「標準化」という手順を使うんだ。各データ点から平均を引いて、それを標準偏差で割るんだ。これによって、新しいデータセットの平均は0、分散は1になるよ。</div>
<img class="hakase-pic" src="pictures\shibainu_4_3.png">
</div>

<div class="speech-container">
<div class="hakase-txt">ちなみに式はこうだよ。</div>
<img class="hakase-pic" src="pictures\shibainu_4_3.png">

<div class="definition-container">
$$
z = \frac{x - \mu}{\sigma}
$$
ただし、$x$は確率変数、$\mu$は平均（期待値）、$\sigma$は標準偏差です。
</div>

<div class="speech-container">
<img class="child-pic" src="pictures\shibainu_2_1.png" alt="子供の顔画像">

<h2>標準化をした正規分布の例</h2>
<div class="speech-container">
<div class="hakase-txt">最後に正規分布を標準化したときの分布を比べてみるね</div>
<img class="hakase-pic" src="pictures\shibainu_4_2.png">
</div>
<div class="speech-container">
<img class="child-pic" src="pictures\shibainu_3_1.png" alt="子供の顔画像">
<p class="child-txt">わあ、標準正規分布って本当に便利なんだね! これからデータを見る時は、標準化のことも考えながら見てみるよ!</p>
</div>

```python
# 平均が1、分散が2の分布のデータを標準化する

# 元の分布のパラメータ
mu_original = 1  # 平均
sigma_original = np.sqrt(2)  # 分散が2のため、標準偏差はsqrt(2)

# 正規分布に従うデータの生成（元の分布）
x_original = np.linspace(mu_original - 10*sigma_original, mu_original + 10*sigma_original, 1000)
y_original = norm.pdf(x_original, mu_original, sigma_original)

# データの標準化
x_standardized = (x_original - mu_original) / sigma_original
y_standardized = norm.pdf(x_standardized, 0, 1)

# 元の分布と標準化された分布のプロット
plt.figure(figsize=(6, 3))

# 元の分布のプロット
plt.plot(x_original, y_original, label='標準化前(μ=1, σ²=2)', color='blue')

# 標準化された分布のプロット
plt.plot(x_standardized, y_standardized, label='標準化後(μ=0, σ²=1)', linestyle='--', color='red')

plt.title('標準化前後の正規分布の比較')
plt.xlabel('X')
plt.xlim(-5, 5)
plt.ylabel('確率密度')
plt.legend()
plt.grid(True)
plt.show()
```
