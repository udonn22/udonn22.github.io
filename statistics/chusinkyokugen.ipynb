<h2>1.サイコロの出目の生成</h2>  
<div class="speech-container">
<div class="hakase-txt">中心極限定理に行く前に、今回の確率分布について理解しよう。</div>
<img class="hakase-pic" src="pictures\shibainu_4_2.png">

<div class="speech-container">
<div class="hakase-txt">10000回、1から6が同じ確率で出るサイコロの出目を生成してみるよ。このような、確率の分布が一定の場合の分布を一様分布と言うよ。</div>

```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib

# サンプル数(サイコロを何回振るか)
n_trials = 10000

# 一様分布からのサンプリング
samples = np.random.randint(1, 7, n_trials)

# ヒストグラムで分布を可視化
plt.figure(figsize=(6, 3))
plt.hist(samples, bins=np.arange(0.5, 7.5), density=True, alpha=0.7)
plt.title(f'1から6までの一様分布で{n_trials}回試行した結果')
plt.xlabel('確率変数X')
plt.ylabel('頻度の割合')
plt.grid(True)
plt.show()
```

<h2>2.中心極限定理の検証</h2>

<div class="speech-container">
<div class="hakase-txt">次に中心極限定理を検証するよ。中心極限定理によれば、$n$回サイコロを投げた時の平均の分布は、$n$が大きくなるにつれて正規分布に近づくはず。</div>
<img class="hakase-pic" src="pictures\shibainu_4_3.png">
</div>

<div class="speech-container">
<div class="hakase-txt">ここでは、サイコロを2回投げた平均$\bar{X}_i$を求める試行を1セットとして、10000セット分の試行をした場合の平均の分布を確認するよ。</div>
<img class="hakase-pic" src="pictures\shibainu_4_2.png">
</div>

<div class="speech-container">
<img class="child-pic" src="pictures\shibainu_1_2.png" alt="子供の顔画像">

</div>

<div class="speech-container">
<div class="hakase-txt">そうだよ。1セットのなかで投げるサイコロの数を2から5、10、100と多くすると、$\bar{X}_i$の分布は正規分布に近づくはずだよ。</div>
<img class="hakase-pic" src="pictures\shibainu_4_2.png">
</div>

```python
# 図を作成する関数を作成
# n_values: 一度の試行でサイコロを振る回数
# n_trials: 試行回数
def plot_histogram(n_values, n_trials):
    # 平均の計算
    means = np.mean(np.random.randint(1, 7, (n_trials, n_values)), axis=1)

    # ヒストグラムで分布を可視化
    plt.figure(figsize=(6, 3))
    plt.hist(means, bins=np.arange(1-1/n_values,6+1/n_values*2,1/n_values), density=True, alpha=0.7)
    plt.title(f'{n_values}回サイコロを投げて平均を算出する試行を{n_trials}回繰り返した結果')
    plt.xlabel(r'各試行での確率変数の平均値$\bar{X}$')
    plt.ylabel('頻度の割合')
    plt.xlim([1-1/n_values, 6+1/n_values*2])
    plt.grid(True)
    plt.show()
```

```python
plot_histogram(n_values=2, n_trials=10000)
```

<h1>中心極限定理の定義</h1>
<div class="speech-container">
<div class="hakase-txt">ちょっと難しいけど中心極限定理の定義はこうだよ。</div>
<img class="hakase-pic" src="pictures\shibainu_4_2.png">
</div>

<div class="definition-container">
ある母集団から無作為に抽出されたサイズ$n$の標本の平均$\bar{X}$は、$n$ が十分に大きい場合に以下の分布に近づく
$$
f \left(x \left|\mu,\frac{\sigma^2}{n}\right) \right.
 = 
 \frac{1}{\sqrt{2\pi (\sigma^2/n)}} 
 e^{-\frac{(x-\mu)^2}{2(\sigma^2/n)}}
$$
ただし、$μ$は母集団の平均、$σ^2$は母集団の分散である。
これは、元の母集団の分布が正規分布でなくても成立する。
</div>


<link rel="stylesheet" type="text/css" href="style.css">
<div class="speech-container">
<div class="hakase-txt">$n$が十分に大きいときは、正規分布を$N$で表すと、nが増えた時に$\bar{X}$は以下の式で近似的に考えることができよ。</div>
<img class="hakase-pic" src="pictures\shibainu_4_2.png">
</div>

$$
\bar{X} \sim N\left(\mu,\frac{\sigma^2}{n}\right)
$$
