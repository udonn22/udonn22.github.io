```python
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np

a = widgets.IntSlider(min=0, max=10, step=1, value=1)
display(a)

def update_plot():
    plt.clf()
    plt.close()
    x = np.linspace(0, 2 * np.pi, 1000)
    y = np.sin(a.value * x)
    plt.plot(x, y)
    plt.show()

a.observe(lambda change: update_plot(), names='value')

update_plot()
```

```python
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import clear_output

a = widgets.IntSlider(min=0, max=10, step=1, value=1)
display(a)

def update_plot(change):
    clear_output(wait=True)
    display(a)
    x = np.linspace(0, 2 * np.pi, 1000)
    y = np.sin(a.value * x)
    plt.plot(x, y)
    plt.show()

a.observe(update_plot, names='value')

update_plot(None)
```
