<h1>カイ二乗分布ってどんな分布?</h1>
<div class="speech-container">

<p class="child-txt">博士、今度はカイ二乗分布っていうのを聞いたんだけど、それは何？</p>
</div>

<div class="speech-container">
<div class="hakase-txt">カイ二乗分布は統計学でよく使われる分布の一つだよ。これは、独立した標準正規分布に従う複数の変数の二乗和が従う分布を指すんだ。</div>

<div class="speech-container">
<img class="child-pic" src="pictures\shibainu_2_2.png" alt="子供の顔画像">
<p class="child-txt">標準正規分布はこの間教えてもらったけど、その二乗和ってどういうこと？</p>
</div>

<div class="speech-container">
<div class="hakase-txt">もちろんだよ。標準正規分布ってのは、平均が0で、標準偏差が1の正規分布のことを言うんだったよね。</div>
<img class="hakase-pic" src="pictures\shibainu_4_2.png">
</div>

<div class="speech-container">
<div class="hakase-txt">そして、その分布に従うランダムな変数をいくつか取って、それぞれを二乗して足し合わせたものが、カイ二乗分布に従うってわけだ。</div>
<img class="hakase-pic" src="pictures\shibainu_4_3.png">
</div>

<div class="speech-container">
<div class="hakase-txt">式で表すとこうだね。</div>
<img class="hakase-pic" src="pictures\shibainu_3_3.png">
</div>

<div class="definition-container">
$$
X_1, X_2, …, X_\nu \sim N\left(0, 1\right) \qquad i.i.d.
$$
のとき、
$$
Y = X_1^2+X_2^2+…+X_\nu^2
$$
の$Y$は自由度$\nu$のカイ二乗分布に従う。
ただし、$i.i.d.$とは、独立同分布(independent and identically distributed)を表す。
</div>

<h1>カイ二乗分布ってどんな形?</h1>
<div class="speech-container">
<img class="child-pic" src="pictures\shibainu_3_2.png" alt="子供の顔画像">
<p class="child-txt">カイ二乗分布ってどんな形になるの?正規分布からできるから、ベル状になるの?</p>
</div>

<div class="speech-container">
<div class="hakase-txt">いいや、カイ二乗分布は二乗した値を足し合わせたものだから、左右対称にならないんだ。実際にプログラミングで見てみよう。</div>
<img class="hakase-pic" src="pictures\shibainu_3_3.png">
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import japanize_matplotlib

# 自由度のリスト
degrees_of_freedom = [2,3,4, 5, 10]

# Yの値
Y = np.linspace(0, 20, 1000)

# Plot each chi-squared distribution
plt.figure(figsize=(6, 4))
for df in degrees_of_freedom:
    plt.plot(Y, stats.chi2.pdf(Y, df), label=f'自由度:{df}')

plt.title('カイ二乗分布の確率密度関数')
plt.xlabel('Y')
plt.ylabel('確率密度')
plt.legend()
plt.grid()
plt.show()
```
