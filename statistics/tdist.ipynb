<h1>t分布の分布形状</h1>
<div class="speech-container">
<div class="hakase-txt">グラフを見る前に、以下のt分布の定義を確認しよう。</div>
<img class="hakase-pic" src="pictures\shibainu_3_3.png">
</div>

<div class="definition-container">
$$U \sim N(0,1),$$
$$V_m \sim \chi^2(m), $$
$$U \text{と} V_m \text{は独立とするとき、}$$
$$T = \frac{U}{\sqrt{V_m/m}} $$
$$\text{は自由度$m$のt分布$t(m)$に従う.}$$
</div>

```python
# Re-import necessary libraries after code execution state reset
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import japanize_matplotlib

# t分布の自由度
degrees_of_freedom_t = [1, 3, 10, 100]

# X values for our plot
x_t = np.linspace(-5, 5, 1000)

# Plot each t-distribution
plt.figure(figsize=(6, 4))
for df in degrees_of_freedom_t:
    plt.plot(x_t, stats.t.pdf(x_t, df), label=f'自由度:{df}')

plt.title('自由度ごとのt分布の確率分布')
plt.xlabel('t')
plt.ylabel('確率分布')
plt.legend()
plt.grid()
plt.show()
```

<h1>t分布と正規分布への外れ値の影響</h1>

<div class="speech-container">
<div class="hakase-txt">外れ値の影響について簡単に見てみよう。平均0、分散1の正規分布からランダムな数字を9つ作って、そこに外れ値として20を追加した、合計10個のサンプルの場合を考えてみるね。</div>
<img class="hakase-pic" src="pictures\shibainu_3_3.png">
</div>

<div class="speech-container">
<div class="hakase-txt">このサンプルから正規分布とt分布の分布を作成したものが以下になるんだ。</div>
<img class="hakase-pic" src="pictures\shibainu_3_3.png">
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# データセットを生成（9つの正常値と1つの異常値）
np.random.seed(0)
normal_values = np.random.normal(loc=0, scale=1, size=9)  # 正常値
outlier = np.array([20])  # 異常値
values_with_outlier = np.append(normal_values, outlier)  # 正常値と異常値を結合

# 平均と標準偏差を計算
mean_outlier = np.mean(values_with_outlier)
std_outlier = np.std(values_with_outlier, ddof=1)

# 自由度（データ数-1）
df = len(values_with_outlier) - 1

# 正規分布の範囲を定義
x_outlier = np.linspace(mean_outlier - 4*std_outlier, mean_outlier + 4*std_outlier, 100)

# t分布の範囲を定義
x_t = np.linspace(-4, 4, 100)

# グラフを描画
plt.figure(figsize=(6, 4))

# 異常値を含む正規分布をプロット
plt.plot(x_outlier, stats.norm.pdf(x_outlier, mean_outlier, std_outlier), label='異常値を含む正規分布', color='blue')

# t分布をプロット
plt.plot(x_t, stats.t.pdf(x_t, df, 0, 1), label='t分布', color='orange')

# サンプルの位置に点をプロット
plt.scatter(values_with_outlier, np.zeros_like(values_with_outlier), color='green', zorder=5, marker='x', label='サンプルの位置')


plt.title('異常値を含む正規分布とt分布の比較')
plt.legend()
plt.grid(True)  # グリッドを表示
plt.show()
```
